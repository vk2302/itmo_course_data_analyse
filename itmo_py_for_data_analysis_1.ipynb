```python
#Создайте класс Chocolate (для определения шоколадного батончика), принимающий следующие аргументы при инициализации: цена батончика (тип - float), добавка к шоколаду (необязательный аргумент, тип - str) - 2 балла
#При инициализации необходимо проверять, чтобы цена батончика была больше нуля, если цена меньше нуля необходимо вызывать исключение ValueError - 2 балла
#Реализуйте метод show_my_chocolate, выводящий на печать "Шоколадный батончик, добавка: {ДОБАВКА}" в случае наличия добавки, а иначе должна отобразиться следующая фраза: "Обычный шоколадный батончик" - 2 балла
#Переопределите метод __str__, который при применении функции print() к объектy класса Chocolate должен возвращать строку "Шоколадный батончик, <ЦЕНА> рублей" - 2 балла
#Продемонстрируйте использование соответствующего функционала вашего класса, создав несколько объектов Chocolate с различными аргументами - 2 балла
```

```python

class Chocolate:
  def __init__(self, price, add=None):
    if price < 0:
      raise ValueError("ValueError")
    self.price = price
    self.add = add

  def show_my_chocolate(self):
    if not self.add:
      print("Обычный шоколадный батончик")
    else:
      print(f"Шоколадный батончик, добавка: {self.add}")
  def __str__(self):
    return f"Шоколадный батончик, {self.price} рублей"

```

```python
ch1 = Chocolate(-1)
ch1.show_my_chocolate()
print(ch1.price)
print(ch1)
```

```python
ch2 = Chocolate(3)
print(ch2.price, ch2.add)
ch2.show_my_chocolate()
print(ch2)
```

```python
ch3 = Chocolate(3, "молоко")
print(ch3.price, ch3.add)
ch3.show_my_chocolate()
print(ch3)
```

##GitHub##

role: connect all users during a project + possibility to go back to earlier version. To edit and improve a code you separate it from a master branch, edit it (commit changes), then put a pull request to have it accepted by readers who can check that the code still works and comment on improvements, before the code is merged back to the branch

master branch: original
main branch: the one we modify

first, it is necessary to create a main branch on web github. Then, on the terminal, you create a folder via **git clone**, and cd to this folder, then create a new branch where you will edit the code from main. this is done via **git checkout -b newbranch**. then manually you create a new folder in this folder and a new file via terminal **touch hwk2.txt**.

then, before commit, you need to add on terminal the new folder + file: **git add hwk2**. then we start to commit (modify the "main" code with a new branch): **git commit -m "add hwk2"**. To make sure that the new branch also appears on the web github page (public profile), we write **git push --set-upstream origin hwk2** in terminal, then enter username and token as password.

We then go to the web profile and check that it has been updated, that the folders and files of the branch are now visible, modify them, and go to the next step - **pull request**: we request for checks to be done by collaborators and other users, before our modifications can be approved and added to the main branch. 1. Settings => Collaborators => add, then go to "Reviewers" on the main page of our code and add reviewers.

Then we wait for feedback and test of the code, and when we get an "approved" from the reviewers, we can merge the branch back with the master branch.
