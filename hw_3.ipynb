```python
# Count non-null values in the 'Name' column
df1 = df[["Name", "Sex"]]
df1
```

```python
df1['firstName'] = df1['Name'].str.split().str[-1]
df1
```

```python


most_common_female = df1[df1['Sex'] == 'female']['firstName'].mode()[0]
most_common_female
```

```python
most_common_male = df1[df1['Sex'] == 'male']['firstName'].mode()[0]
most_common_male
```

```python
df3 = df[["Name", "Sex", "Pclass"]]
df3['firstName'] = df1['firstName']
display(df3.groupby(["Sex", "Pclass"])["firstName"].value_counts())
```

```python
most_common_female_pclass1 = df3[(df3['Sex'] == 'female') & (df3['Pclass'] == 1)]['firstName'].mode()[0]
display(most_common_female_pclass1)
most_common_male_pclass1 = df3[(df3['Sex'] == 'male') & (df3['Pclass'] == 1)]['firstName'].mode()[0]
display(most_common_male_pclass1)
most_common_female_pclass2 = df3[(df3['Sex'] == 'female') & (df3['Pclass'] == 2)]['firstName'].mode()[0]
display(most_common_female_pclass2)
most_common_male_pclass2 = df3[(df3['Sex'] == 'male') & (df3['Pclass'] == 2)]['firstName'].mode()[0]
display(most_common_male_pclass2)
most_common_female_pclass3 = df3[(df3['Sex'] == 'female') & (df3['Pclass'] == 3)]['firstName'].mode()[0]
display(most_common_female_pclass3)
most_common_male_pclass3 = df3[(df3['Sex'] == 'male') & (df3['Pclass'] == 3)]['firstName'].mode()[0]
display(most_common_male_pclass3)
```
